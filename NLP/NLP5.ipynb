```python
import nltk
import random
import string
from itertools import combinations

nltk.download('words')
nltk.download('brown')
nltk.download('stopwords')

from nltk.corpus import words,brown,stopwords
english_words = set(w.lower() for w in words.words())
```

```python
random_words = random.sample(list(english_words),10)
print("10 Random Words: " ,random_words)
```

```python
def isValid(word):
  return word.lower() in english_words

user_word = input("Enter a word to check: ")

if isValid(user_word):
  print("Valid!!!")
else:
  print("Invalid!!!")
```

```python
def valid_combos(word,k):
  word = word.lower()
  combo = combinations(word,k)
  valid_words = set()

  for c in combo:
    new_word = ''.join(c)
    if new_word in english_words:
      valid_words.add(new_word)
  print(f"Valid combinations of length {k}: ")
  for w in valid_words:
    print(w)
msg = input("Enter a word: ")
k = int(input("Enter the value of K: "))
valid_combos(msg,k)
```

```python
brown_words = [w.lower() for w in brown.words() if w.isalpha()]
sample_words = random.sample(brown_words,100)

prefixes = ["un","re","pre","dis"]
suffixes = ["ing","ed","ly","er"]
result = []

for word in sample_words:
  for p in prefixes:
    new_word = p+word
    if isValid(new_word):
      result.append(new_word)
  for s in suffixes:
    new_word = word+s
    if isValid(new_word):
      result.append(new_word)
print("Valid generated words: ",result)
```

```python
stop_words = set(stopwords.words('english'))
clean_words = []
for w in brown.words():
  word = w.lower()
  if word.isalpha() and word not in stop_words:
    clean_words.append(word)
unique_words = set(clean_words)
total_length = 0
for w in unique_words:
  total_length += len(w)

avg_length = total_length/len(unique_words)

print("Total unique words: ",len(unique_words))
print("Average word length: ",avg_length)
```
