```python
import re
import nltk
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
text = input("Enter the text: ").lower()
print("\nLower case Text: ",text)
```

```python
text = re.sub(r'<.*?>','',text)
print("\nAfter removing HTML tags:\n",text)
```

```python
text = re.sub(r'[^a-zA-Z\s]','',text)
print("\nAfter Script Validation:\n",text)
```

```python
tokens = word_tokenize(text)
print("\nTokens: ",tokens)
```

```python
frequency = Counter(tokens)
print("\nWord frequency:\n", frequency)
print("\nNumber of Unique Words: ",len(frequency))
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]
print("\nAfter stopword removal: \n",filtered_tokens)
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("\nAfter stemming:",stemmed_words)
```
