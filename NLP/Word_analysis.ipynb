```python
import nltk,string
from collections import Counter,defaultdict
from nltk.corpus import brown,stopwords
nltk.download('brown')
nltk.download('stopwords')
```

```python
print(brown.words()[:10])
```

```python
print("Sentence")
l =  brown.sents()[:10]
for i in range(len(l)):
    print("Sentence ",i,"\n",l[i],"\n") 
```

```python
words_Without_Punctuation = []
words = brown.words()
print("Words with puntuation length:",len(words))
for word in words:
    if word not in string.punctuation:
        words_Without_Punctuation.append(word)
print(words_Without_Punctuation[:20])
print("Words without punctuation length:",len(words_Without_Punctuation))
```

```python
stop_words = set(stopwords.words('english'))
filtered_Words = []
for word in words:
    if word.isalpha() and word not in stop_words:
        filtered_Words.append(word.lower())

print("Words length after removing stop words:",len(filtered_Words))
```

```python
vocabulary = set(filtered_Words)
vocabulary = sorted(vocabulary)
print("Length of sorted Vocabulary: ",len(vocabulary))
```

```python
frequency_Dictionary = {}
words_1 = brown.words()[:100]
filtered_words_1 = [word for word in words_1 if word in vocabulary]
frequency_Dictionary = Counter(filtered_words_1)
print(frequency_Dictionary)
```

```python
print(string.ascii_lowercase)
sample_word = defaultdict(list)
for char in string.ascii_lowercase:
    for word in vocabulary:
        if word.startswith(char):
            if char in sample_word:
                if(len(sample_word[char])<3):
                    sample_word[char].append(word)
            else:
                sample_word[char]
for i in sample_word:
    print(i,sample_word[i])
```
