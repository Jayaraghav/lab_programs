```python
def edit_distance_with_operations(s1, s2):
    m, n = len(s1), len(s2)
    # DP table: dp[i][j] = minimum edit distance between s1[:i] and s2[:j]
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    # To store operations
    op = [[[] for _ in range(n + 1)] for _ in range(m + 1)]
    for i in range(m + 1):
        dp[i][0] = i
        if i > 0:
            op[i][0] = op[i - 1][0] + [f"Delete '{s1[i-1]}' from position {i-1}"]
    for j in range(n + 1):
        dp[0][j] = j
        if j > 0:
            op[0][j] = op[0][j - 1] + [f"Insert '{s2[j-1]}' at position {j-1}"]
    # Fill DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
                op[i][j] = op[i - 1][j - 1][:]

            else:
                insert_cost = dp[i][j - 1] + 1
                delete_cost = dp[i - 1][j] + 1
                replace_cost = dp[i - 1][j - 1] + 1

                min_cost = min(insert_cost, delete_cost, replace_cost)
                if min_cost == insert_cost:
                    dp[i][j] = insert_cost
                    op[i][j] = op[i][j - 1] + [f"Insert '{s2[j-1]}' at position {i}"]
                elif min_cost == delete_cost:
                    dp[i][j] = delete_cost
                    op[i][j] = op[i - 1][j] + [f"Delete '{s1[i-1]}' from position {i-1}"]
                else:
                    dp[i][j] = replace_cost
                    op[i][j] = op[i - 1][j - 1] + [f"Substitute '{s1[i-1]}' with '{s2[j-1]}' at position {i-1}"]

    return dp[m][n], op[m][n]

if __name__ == "__main__":
    try:
        # Get input from user
        word1 = input("Enter the first word: ").strip()
        word2 = input("Enter the second word: ").strip()
        if not word1.isalpha() or not word2.isalpha():
            raise ValueError("Please enter valid words containing only letters.")

        distance, operations = edit_distance_with_operations(word1, word2)
        print(f"\nEdit Distance between '{word1}' and '{word2}': {distance}")
        print("Operations to transform:")
        if operations:
            for step, op in enumerate(operations, start=1):
                print(f"{step}. {op}")
        else:
            print("No operations needed. Words are identical.")
    except Exception as e:
        print(f"Error: {e}")
```
