```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
text = input("Enter a sentence: ")
result = nlp(text)
```

```python
print("Word\t | Leema\t| POS\t | Morphological Features")
print("-"*70)

for token in result:
  person = token.morph.get("Person",[""])[0]
  tense = token.morph.get("Tense",[""])[0]
  number = token.morph.get("Number",[""])[0]

  features = f"Person:{person or 'N/A'},Tense: {tense or 'N/A'}, Number :{number or 'N/A'}"

  print(f"{token.text:10} | {token.lemma_:10} | {token.pos_:6} | {features}")
```
