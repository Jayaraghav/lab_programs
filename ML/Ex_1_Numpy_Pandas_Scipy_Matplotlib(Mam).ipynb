#Program using scikit learn for performing ML tasks
1. Load the Iris dataset from `sklearn.datasets`
2. Split the data into training and testing sets
3. Perform feature scaling using `StandardScaler`
(150 samples of 3 iris species, 4 features each).

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data      # Features, Extracts the feature matrix (shape: (150, 4)).
y = iris.target    # Labels, Extracts the target labels (0, 1, 2 for the three species).
print(X.shape)
print(y.shape)
print(iris.feature_names) #sepal length, sepal width, petal length and petal width
print(iris.target_names) #setosa, virginica, versicolor

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

#X_train → Features for training the model
#X_test → Features for testing the model
#y_train → Labels for training
#y_test → Labels for testing
#he model learns patterns from X_train → y_train.
Then we can check accuracy on X_test → compare predictions with y_test

X_train.shape  # (105, 4) → 70% of 150
X_test.shape   # (45, 4)  → 30% of 150
y_train.shape  # (105,)
y_test.shape   # (45,)


# Feature scaling
scaler = StandardScaler() #create scalar object
X_train_scaled = scaler.fit_transform(X_train) #fit->mean and std dev,
X_test_scaled = scaler.transform(X_test) #tranform->scales each feature
print(X_train_scaled)
print(X_test_scaled)
```

```python
import numpy as np
from scipy.optimize import fsolve

# Define the equation
def equation(x):
    return x**3 - 5*x + 3

# Initial guess
initial_guess = 0

# Solve
root = fsolve(equation, initial_guess)

print("Root of the equation:", root)
```
