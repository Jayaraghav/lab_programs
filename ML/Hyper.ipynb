```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, HalvingGridSearchCV
from sklearn.ensemble import RandomForestClassifier
from skopt import BayesSearchCV
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
print(df.head())
```

```python
X = df.drop('target', axis=1)
y = df['target']
print("Dependent Features (X):")
print(X.head())
print("Independent Features (y):")
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}
```

```python
manual_rf = RandomForestClassifier(n_estimators=10, max_depth=5, random_state=42)
manual_rf.fit(X_train, y_train)
y_pred = manual_rf.predict(X_test)
manual_acc = accuracy_score(y_test, y_pred)
```

```python
grid = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3)
grid.fit(X_train, y_train)
y_pred = grid.predict(X_test)
grid_acc = accuracy_score(y_test, y_pred)
```

```python
random_s = RandomizedSearchCV(RandomForestClassifier(random_state=42), param_grid, n_iter=10, cv=3, random_state=42)
random_s.fit(X_train, y_train)
y_pred = random_s.predict(X_test)
random_acc = accuracy_score(y_test, y_pred)
```

```python
halving = HalvingGridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, factor=2, random_state=42)
halving.fit(X_train, y_train)
y_pred = halving.predict(X_test)
halving_acc = accuracy_score(y_test, y_pred)
```

```python
bayesian = BayesSearchCV(RandomForestClassifier(random_state=42), param_grid, n_iter=10, cv=3, random_state=42)
bayesian.fit(X_train, y_train)
y_pred = bayesian.predict(X_test)
bayesian_acc = accuracy_score(y_test, y_pred)
```

```python
print("Manual Search:", manual_acc)
print("Grid Search:", grid_acc)
print("Randomized Search:", random_acc)
print("Halving Search:", halving_acc)
print("Bayesian Search:", bayesian_acc)
```
