```python
import pandas as pd
dic = {
    "Order_no":[101,104,105,110,115,113,118,134,124,150,159],
    "Amount":[1345,2345,9875,430,235,568,2390,9560,3490,3409,905],
    "Customer_id":["C1","C1","C1","C1","C1","C1","C1","C1","C1","C1","C2"]
}
df = pd.DataFrame(dic)
print(df)
```

```python
print("Mean = \n",df.groupby("Customer_id")["Amount"].mean()," \nMaximum = \n",df.groupby("Customer_id")["Amount"].max()," \nMinimum = \n",df.groupby("Customer_id")["Amount"].min())
```

```python
print("Groups = \n",df.groupby("Customer_id").count())
```
