```python
from scipy.optimize import fsolve
def f(x):
    return x**3 - 5*x + 3
roots = [
    fsolve(f, -3)[0],
    fsolve(f,  0)[0],
    fsolve(f,  3)[0]
]
print("Roots:", roots)
```

```python
from scipy.integrate import quad
def f(x):
    return x**2 + 2*x
result, error = quad(f, 0, 4)

print("Integral value:", result)
print("Estimated error:", error)
```
