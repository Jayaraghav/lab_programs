```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data={
    'products':['A','B','C','D','E','F','G','H'],
    'sales':[150,60,55,25,160,85,70,50],
    'revenue':[1500000,70000,49000,38000,76500,1000000,300000,500000],
    'ratings':[4.7,4.3,4.6,3.5,4.6,4.0,3.9,4.5]
    }
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df['products'],df['sales'],marker="o")
plt.xlabel('Products')
plt.ylabel('Sales (Units)')
plt.title('Sales Trends')
plt.show()
```

```python
plt.bar(df['sales'],df['revenue'])
plt.xlabel('Sales')
plt.ylabel('Revenue (Rs)')
plt.title('Revenue by Product')
plt.show()
```

```python
plt.scatter(df['sales'],df['ratings'])
plt.xlabel('Sales (Units)')
plt.ylabel('Customer Ratings')
plt.title('Sales vs Customer Ratings')
plt.show()
```

```python
plt.hist(df['sales'],bins=5)
plt.xlabel('Sales (Units)')
plt.ylabel('Frequency')
plt.title('Distribution of Sales')
plt.show()
```

```python
plt.pie(df['sales'],labels=df['products'],autopct='%1.1f%%')
plt.title('Sales Distribution by Product')
plt.show()
```
